```python
!pip install transformers diffusers accelerate sentencepiece torch --quiet

from transformers import pipeline
from diffusers import StableDiffusionPipeline
from IPython.display import display, HTML
import torch
import re
import requests

input_text = """
Photosynthesis is a process used by plants and other organisms to convert light energy into chemical energy.
This energy is stored in carbohydrate molecules, such as sugars, which are synthesized from carbon dioxide and water.
Photosynthesis is essential for life on Earth.
"""

simplifier = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
def simplify_text(text):
    simplified = simplifier(text, max_length=60, min_length=25, do_sample=False)
    return simplified[0]['summary_text']

print("🔍 Simplified Explanation:\n")
print(simplify_text(input_text))

def make_flashcards(text):
    sentences = re.split(r'\.\s+', text)
    questions = []
    for s in sentences:
        if " is " in s:
            parts = s.split(" is ", 1)
            questions.append(f"Q: What is {parts[0]}?")
            questions.append(f"A: {parts[1]}")
    return questions

print("\n🧠 Quiz Questions:\n")
for q in make_flashcards(input_text):
    print(q)

def show_craiyon_link(query):
    search_query = query.replace(" ", "+")
    link = f"https://www.craiyon.com/?prompt={search_query}"
    display(HTML(f"<br><b>🖼️ View Diagram on Craiyon:</b> <a href='{link}' target='_blank'>{link}</a>"))

def generate_diagram_sd(prompt):
    if not torch.cuda.is_available():
        print("⚠️ GPU not available. Cannot generate Stable Diffusion image.")
        return

    print("⏳ Generating image using Stable Diffusion...")
    pipe = StableDiffusionPipeline.from_pretrained(
        "CompVis/stable-diffusion-v1-4",
        torch_dtype=torch.float16
    ).to("cuda")

    image = pipe(prompt).images[0]
    display(image)

diagram_prompt = "educational diagram of photosynthesis in plants"

print("\n--- OPTION 1: View with Craiyon ---")
show_craiyon_link(diagram_prompt)

print("\n--- OPTION 2: Generate with Stable Diffusion (requires Colab GPU) ---")
generate_diagram_sd(diagram_prompt) #Make sure GPU is enabled if using Option 2 → Runtime > Change runtime type > GPU
```

```python
!pip install transformers diffusers accelerate sentencepiece torch scikit-learn --quiet

from transformers import pipeline
from diffusers import StableDiffusionPipeline
from IPython.display import display, HTML
import torch
import re
import requests
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

input_text = """
The water cycle, also known as the hydrological cycle, describes how water moves around the Earth. It is a continuous process involving the movement of water between the atmosphere, land, and oceans.

Water from oceans, lakes, and rivers evaporates due to the heat from the sun. This vapor rises into the atmosphere.

Plants also release water vapor through transpiration, where water moves from roots to leaves and evaporates.

The water vapor cools as it rises, condensing into tiny droplets that form clouds.

When these droplets combine and grow heavy, they fall to the Earth as precipitation, such as rain, snow, sleet, or hail.

Some precipitation flows over the land as surface runoff, entering rivers, lakes, and oceans.

Other precipitation infiltrates the soil, replenishing groundwater supplies stored in aquifers.

Groundwater can return to the surface through springs or seep into water bodies.

The cycle repeats as water evaporates again, maintaining the Earth's water balance.

The water cycle is essential for distributing fresh water and sustaining ecosystems.

Human activities like deforestation and urbanization can disrupt the cycle.

Climate change also affects evaporation and precipitation patterns.

Understanding the water cycle helps manage water resources and predict weather.

It plays a vital role in agriculture, industry, and drinking water supply.

The water cycle demonstrates the interconnectedness of Earth's systems.
"""

simplifier = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
def simplify_text(text):
    simplified = simplifier(text, max_length=60, min_length=25, do_sample=False)
    return simplified[0]['summary_text']

print("🔍 Simplified Explanation:\n")
print(simplify_text(input_text))

def make_flashcards(text):
    sentences = re.split(r'\.\s+', text)
    questions = []
    for s in sentences:
        if " is " in s:
            parts = s.split(" is ", 1)
            if len(parts) == 2:
                questions.append(f"Q: What is {parts[0]}?")
                questions.append(f"A: {parts[1]}")
    return questions

print("\n Quiz Questions:\n")
for q in make_flashcards(input_text):
    print(q)

def extract_keywords(text, top_n=5):
    vectorizer = CountVectorizer(stop_words='english', ngram_range=(1, 2))
    X = vectorizer.fit_transform([text])
    freqs = np.asarray(X.sum(axis=0)).flatten()
    terms = vectorizer.get_feature_names_out()
    sorted_terms = [term for _, term in sorted(zip(freqs, terms), reverse=True)]
    return sorted_terms[:top_n]

keywords = extract_keywords(input_text, top_n=5)
concept_string = ", ".join(keywords)
diagram_prompt = f"educational diagram showing {concept_string}"

print("\n Auto-generated Diagram Prompt:\n")
print(diagram_prompt)

def show_craiyon_link(query):
    search_query = query.replace(" ", "+")
    link = f"https://www.craiyon.com/?prompt={search_query}"
    display(HTML(f"<br><b>🖼️ View Diagram on Craiyon:</b> <a href='{link}' target='_blank'>{link}</a>"))

def generate_diagram_sd(prompt):
    if not torch.cuda.is_available():
        print("⚠️ GPU not available. Cannot generate Stable Diffusion image.")
        return

    print("⏳ Generating image using Stable Diffusion...")
    pipe = StableDiffusionPipeline.from_pretrained(
        "CompVis/stable-diffusion-v1-4",
        torch_dtype=torch.float16
    ).to("cuda")

    image = pipe(prompt).images[0]
    display(image)

print("\n--- OPTION 1: View with Craiyon ---")
show_craiyon_link(diagram_prompt)

print("\n--- OPTION 2: Generate with Stable Diffusion (requires GPU) ---")
generate_diagram_sd(diagram_prompt)
```
